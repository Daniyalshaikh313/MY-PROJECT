```python
Top_10_City	= df['City'].value_counts()
Top_10_City
fig = px.bar(Top_10_City, x=Top_10_City.index, y=Top_10_City.values, title='Top 10 City with Most Restaurants')

fig.update_xaxes(title_text='City', title_font_color='white')  
fig.update_yaxes(title_text='Number of Restaurants', title_font_color='white')  

fig.update_layout(
    paper_bgcolor='black', 
    plot_bgcolor='black',
    font_color='white'
)

fig.show()    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
ax = sns.histplot(data=df, x='Price', kde=False, color='blue')

ax.set_title('Distribution of Restaurant Prices on Swiggy', color='white', fontsize=16)
ax.set_xlabel('Price', color='white', fontsize=12)
ax.set_ylabel('Count', color='white', fontsize=12)

ax.set_facecolor('black') 
plt.gcf().set_facecolor('black')  
ax.tick_params(colors='white') 
ax.grid(color='gray', linestyle='--', linewidth=0.5)  

plt.tight_layout()
plt.show()
```

```python
fig = px.scatter(df, x='Price', y='Avg ratings', title='Correlation Between Restaurant Price and average ratings', color_discrete_sequence=['red'])
fig.update_layout(template="plotly_dark") 

fig.show()
```

```python
Top_rating_Resturant_in_Mumbai = df[(df['City']=='Mumbai')&(df['Avg ratings']==5.0)]
Top_rating_Resturant_in_Mumbai
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
ax = sns.barplot(
    data=Top_rating_Resturant_in_Mumbai, 
    x='Restaurant', 
    y='Avg ratings', 
    palette='Reds_r'  
)

ax.set_title('Top-Rated Restaurants in Mumbai', color='white', fontsize=18, fontweight='bold')
ax.set_xlabel('Restaurant', color='white', fontsize=12, labelpad=8)
ax.set_ylabel('Average Ratings', color='white', fontsize=12, labelpad=8)


ax.set_xticklabels(ax.get_xticklabels(), rotation=40, horizontalalignment='right', color='white', fontsize=12)

ax.set_facecolor('black')  
plt.gcf().set_facecolor('black') 
ax.tick_params(colors='white')  
ax.grid(color='gray', linestyle='--', linewidth=0.3)  


for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='edge', color='white', fontsize=10)

plt.tight_layout()
plt.show()
```
